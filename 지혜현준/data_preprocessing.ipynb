```python
import pandas as pd
```

```python
# Read the Excel file into a DataFrame
file_path = 'data/CB Insights AI 100 2024.xlsx'
excel_file = pd.ExcelFile(file_path)
excel_file.sheet_names
```

```python
# Load the main sheet which appears to be '2024 AI 100'
df = pd.read_excel(file_path, sheet_name='2024 AI 100')

# Display the first few rows to identify any header or extra explanation rows
df.head(10)
```

```python
# Use the actual header row (row 5) to set column names and filter the data accordingly
df.columns = df.iloc[5]
df = df[8:]  # Start from actual data rows (row 8 based on preview)

# Clean up the DataFrame by resetting the index and dropping any unnecessary columns
df.reset_index(drop=True, inplace=True)
```

```python

df.head(10)
```

```python
# Drop columns with all NaN values
df = df.dropna(axis=1, how='all')

# Display the updated DataFrame
df.head()
```

```python
# Check for rows where the "Latest Funding Round" is Seed VC or Pre-seed
filtered_df = df[
    df["Latest Funding Round"].str.lower().isin(["seed vc", "pre-seed", "incubator", "convertible note", "grant"])
]
```

```python
filtered_df.head(10)
```

```python
filtered_df.reset_index(drop=True, inplace=True)
filtered_df.to_csv('angel_filtered_data.csv', index=False, header=True)
```
